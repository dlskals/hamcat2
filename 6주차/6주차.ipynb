```python
import requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import re
from gensim.summarization import summarize

search = input()
url = f"https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query=영화+{search}+정보"
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 6.3; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/44.0.2403.157 Safari/537.36"}

res = requests.get(url, headers = headers)
soup = BeautifulSoup(res.text, 'lxml')

sen_t = soup.select_one('div.cm_content_wrap > div.cm_content_area._cm_content_area_synopsis > div > div.intro_box._content > p').text
s = re.sub(r'[^\w\s]', '\n', sen_t)
#s = ''
#for word in words:
    #s += (word + ' ')
words_list = s.split()

sen = []
count = 0
s = ""
for i in words_list:
    count += 1
    if count % 6 == 1 and count != 1:
        sen.append(str(s))
        s = ""
    if count % 6 == 0 and count != 1:
        s += i + "\n "
    else :
        s += i + " "
sen.append(str(s))
sentence = ""

for _ in range(1):
    for i in sen:
        sentence += i
a = sen_t
b = re.sub(r'[.,]', '\n', a)

v1 = summarize(sen_t, ratio=0.3).strip()
v2 = summarize(sentence, ratio=0.3).strip()
v3 = summarize(b, ratio=0.3).strip()

print("\n결과가 있는 경우 중 가장 나은 문장\n")
if v1 != "":
    print("바로 요약 :", v1)
elif v3 !=  "":
    print("따옴표, 쉼표 줄바꿈으로 바꾼 후 요약 :\n", v3)
else:
    print("공백으로 나눈 후 6개 단어씩 문장으로 만든 후 요약 :\n", v2)

print("\n\n경우마다 나온 결과\n")

print("바로 요약 :", v1, '\n')

print("공백으로 나눈 후 6개 단어씩 문장으로 만든 후 요약 :\n", v2, '\n')

print("따옴표, 쉼표 줄바꿈으로 바꾼 후 요약 :", v3)
```
